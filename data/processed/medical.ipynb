```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

# Medikal Türkçe QA için daha uygun model örneği (örnek model adı)
model_name = "kaixkhazaki/turkish-medical-question-answering"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)

# Örnek kullanım
context = "Reflü hastalığında genellikle mide yanması ve yemek sonrası rahatsızlık hissi görülür."
question = "Reflü hastalığı belirtileri nelerdir?"

result = qa_pipeline(question=question, context=context)
print(f"Cevap: {result['answer']}")
```

```python
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)

test_samples = [
    {
        "context": "Reflü hastalığında genellikle mide yanması ve yemek sonrası rahatsızlık hissi görülür.",
        "question": "Reflü hastalığı belirtileri nelerdir?"
    },
    {
        "context": "Dengue hastalığında genellikle yüksek ateş, kas ağrıları ve şiddetli baş ağrısı olur.",
        "question": "Dengue hastalığının belirtileri nelerdir?"
    },
    {
        "context": "Hipotiroidizmde stres azaltılmalı, düzenli egzersiz yapılmalı ve sağlıklı beslenilmelidir.",
        "question": "Hipotiroidizm için hangi önlemler alınmalıdır?"
    },
    {
        "context": "Migren hastalarında meditasyon yapmak, stresi azaltmak ve güneşte polarize gözlük kullanmak faydalıdır.",
        "question": "Migren tedavisinde hangi yöntemler önerilir?"
    },
    {
        "context": "Pneumonia için doktor kontrolü, ilaç tedavisi, dinlenme ve takip önerilir.",
        "question": "Pnömoni tedavisi nasıl yapılır?"
    }
]

for sample in test_samples:
    result = qa_pipeline(question=sample["question"], context=sample["context"])
    print(f"Soru: {sample['question']}")
    print(f"Cevap: {result['answer']}\n")
```

```python
# Örnek veriler
corpus = [
    "Reflü hastalığında genellikle mide yanması ve yemek sonrası rahatsızlık hissi görülür.",
    "Dengue hastalığında yüksek ateş, kas ağrıları ve şiddetli baş ağrısı olur.",
    "Hipotiroidizmde stres azaltılmalı, düzenli egzersiz yapılmalı ve sağlıklı beslenilmelidir.",
    "Migren için meditasyon yapmak, stresi azaltmak ve güneşte polarize gözlük kullanmak önerilir.",
    "Pnömoni tedavisi için doktor kontrolü, ilaç tedavisi, dinlenme ve takip gerekir."
]

```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Embedder yükle
embedder = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

# Cümleleri vektörleştir
embeddings = embedder.encode(corpus, convert_to_numpy=True)

# FAISS index oluştur
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
```
